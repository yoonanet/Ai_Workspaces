```python
from tensorflow import keras
from tensorflow.keras.models import Sequential

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.layers import LSTM, Dense

from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
# 데이터 불러오기
df = pd.read_csv('project2_data/df.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# 정규화하기
# MinMaxScaler 클래스 생성
scaler = MinMaxScaler()

# MinMaxScaler 클래스의 fit_transform() 메서드 사용 데이터 정규화
normalized_data = scaler.fit_transform(df)

# 정규화된 데이터, 데이터 프레임으로 다시 변환
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
```

```python
normalized_df.head()
```

```python
ndf = normalized_df.copy()
```

```python
# 데이터 나누기
X = ndf.iloc[:, :-1]
y = ndf.iloc[:,[-1]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
pd.DataFrame(X_test).to_csv('project2_data/X_test.csv', index=False)
pd.DataFrame(y_test).to_csv('project2_data/y_test.csv', index=False)
```

```python
X_test = pd.read_csv('project2_data/X_test.csv')
y_test = pd.read_csv('project2_data/y_test.csv')
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
from keras.layers import Dropout

model = Sequential()

model.add(LSTM(units = 128, input_shape=(ndf.shape[1] - 1, 1), activation='relu'))
#lstm이 처음에는 3차원으로 데이터를 받아지는데 2차원으로 받을 수 있도록 고정 시킴
#return_sequences는 다음 LSTM 층에 시퀀스 전체를 전달할지 여부를 설정하는 것이다. - true지정 

#model.add(LSTM(units = 64, return_sequences = True))
#Dropout은 신경망의 과적합(overfitting)을 줄이는 데 도움을 주는 정규화(regularization) 기법 중 하나이다.

model.add(Dense(units=64))
#model.add(Dense(units=32))

# Adding the output layer
model.add(Dense(units = 1))

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='mse')

model.summary()

model.fit(X_train, y_train, epochs=25, batch_size=32, verbose=1) # batch_size: 데이터 집합 사이즈 / verbose: 출력되는 정보의 양
#loss: 0.0043 - 위코드대로 했을 때
#rmse : 0.004151275376514095
#r2 : 0.8808119488671571
#mae : 0.04230644207513197


#dense 2개 추가 => 64, 32!!
#loss : 0.0043 
#rmse : 0.004170491872854569 낮을수록 좋음
#r2 : 0.8802602204124802 1에 가까울수록 좋음
#mae : 0.04210032104240598 낮을수록 좋음


# LSTM을 256으로 dense는 128 지정
# loss: 0.0040
#rmse : 0.004146556377860793
#r2 : 0.880947436928481
#mae : 0.04228925071835883

# LSTM을 128, dense는 64, 옵티마이저를 RMSprop로 지정
#RMSprop (Root Mean Square Propagation): RMSprop은 기울기의 제곱값의 이동 평균을 사용하여 학습률을 조정하는 옵티마이저이다. 
# loss: 0.0043
#rmse : 0.004286781657851175
#r2 : 0.8769702592313798
#mae : 0.0424483836106407

# LSTM을 128, dense는 64, 옵티마이저를 Nadam으로 적용
#Nadam은 Adam 옵티마이저와 비슷하지만, Nadam은 추가적으로 Nesterov 모멘텀을 사용하여 가중치 업데이트를 수행합니다.
#loss: 0.0043
#rmse : 0.0042956073264975945
#r2 : 0.8767169643793572
#mae : 0.042578087867107504

# Nadam에서 dense 2개(64, 32)
#loss: 0.0041
#rmse : 0.0041985463139919766
#r2 : 0.8795025952232878
#mae : 0.04234893591044282

#에폭스 여기서부터 40
#병환님 코드 + 재은님이 다시 보내주신 데이터로 돌림
#rmse : 0.0033472387605986563
#r2 : 0.90475347178065
#mae : 0.035972331743353736

#기존 코드에 danse_32 추가
#loss: 0.0031
#rmse : 0.0034113338929678072
#r2 : 0.9022596912231469
#mae : 0.03597691820271569
```

```python
from keras.layers import Dropout

model = Sequential()

model.add(LSTM(units = 128, input_shape=(ndf.shape[1] - 1, 1)))
model.add(Dense(units=32))
#model.add(Dense(units=32))

# Adding the output layer
model.add(Dense(units = 1))

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='mse')

model.summary()
```

```python
model.fit(X_train, y_train, epochs=40, batch_size=32, verbose=1)
```

```python
pred = model.predict(X_test)
print(y_test.shape, pred.shape)
```

```python
# 실제값과 예측값을 계약년월별 평균값으로 계산
predictions = np.array(pred)
actual_values = np.array(y_test)
```

```python
contract_year_predictions = np.array([predictions[i:i + 12].mean() for i in range(0, len(predictions), 12)])
contract_year_actual_values = np.array([actual_values[i:i + 12].mean() for i in range(0, len(actual_values), 12)])
```

```python
# RMSE를 계산(낮을 수록 정확함)
rmse = mean_squared_error(y_test, predictions)
print('rmse :',rmse)

# R^2을 계산(R^2가 1에 가까울수록 모델이 데이터를 잘 설명)
r2 = r2_score(y_test, predictions)
print('r2 :',r2)

# MAE를 계산(MAE가 낮을수록 예측값과 실제값이 가까워짐)
mae = mean_absolute_error(y_test, predictions)
print('mae :',mae)

#rmse : 0.004324419340522558
#r2 : 0.8769708394595936
#mae : 0.0434767911808414
```

```python
plt.plot(contract_year_predictions, label="Predictions", linestyle='solid')
plt.plot(contract_year_actual_values, label="Actual Values")
plt.xlabel("Contract Year")
plt.ylabel("Value")
plt.legend()
plt.show()
```
