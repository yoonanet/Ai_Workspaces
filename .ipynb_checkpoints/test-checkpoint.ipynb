```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('input_202001.csv', names=['동', '면적', '층수', '건축년도', '보증금'], encoding='euc-kr')
data
```

```python
sum = 0
data11 = np.array([sum+=i for i in data['동']])
data11
print(data11)
```

```python
data1 = data[['면적', '보증금']]
data1
```

```python
data1['면적'] = np.around(data1['면적'])
data1
```

```python
data2 = data1.groupby(['면적']).mean().reset_index()
data2
```

```python
X = data2['보증금']
y = data2['면적']

#면적별 보증금 평균값의 산점도 출력
import matplotlib.pyplot as plt
plt.scatter(X, y, alpha=0.4)
plt.title('scatterplot')
plt.xlabel('deposit')
plt.ylabel('area')
plt.show()
```

```python
data3 = data[['층수', '보증금']]
data3
```

```python
data3 = data3.groupby(['층수']).mean().reset_index()
data3
```

```python
X1 = data['보증금']
y1 = data['층수']

#층수별 보증금 산점도
import matplotlib.pyplot as plt
plt.scatter(X1, y1, alpha=0.4)
plt.title('scatterplot')
plt.xlabel('deposit')
plt.ylabel('floor')
plt.show()
```

```python
data4 = data[['건축년도', '보증금']]

data4 = data4.groupby(['건축년도']).mean().reset_index()
data4
```

```python
X = data4['건축년도']
y = data4['보증금']

#건축년도별 보증금 평균값의 산점도 출력
import matplotlib.pyplot as plt
plt.scatter(X, y, alpha=0.4)
plt.title('scatterplot')
plt.xlabel('construction')
plt.ylabel('deposit')
plt.show()
```

```python
data5 = data[['동', '보증금']]

data5 = data5.groupby(['동']).mean().reset_index()
data5
```

```python
X = data5['보증금']
y = data5['동']

#동별 보증금 평균값의 산점도 출력
import matplotlib.pyplot as plt
plt.scatter(X, y, alpha=0.4)
plt.title('scatterplot')
plt.xlabel('deposit')
plt.ylabel('dong')
plt.show()
```
