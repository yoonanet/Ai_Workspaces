```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# 단순한 그래프 그리기

# 데이터 준비
x = np.arange(0, 6, 0.1)
print(x) #0.1씩 증가하면서 최종적으로는 5.9까지 리턴함 => 간격이 0.1임
```

```python
y1 = np.sin(x) #사인함수의 기능 제공
y2 = np.cos(x) #코사인함수의 기능 제공
```

```python
# 그래프 그리기

plt.plot(x, y1, label="sin") #label: 범례의 이름을 지정해주는 것
plt.plot(x, y2, linestyle="--", label="cos") #linestyle: 분포도를 선으로 연결해서 보여주는 것으로 화면에 보여줄 선의 모양을 넣어주면 됨
# cos() 함수는 점선(--)으로 보여주도록 지정한 것임
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title("sin & cos") #시각화되는 화면에 붙여줄 제목 출력
plt.legend() # 범례
plt.show()
```

```python
np.random.seed(1234) # 난수를 고정.

x = np.arange(10)
y = np.random.rand(10)

plt.plot(x, y) # 두 값사이에 분포도를 시각화 -> 기본 디폴트는 꺾은 선 그래프를 등록
#값들의 연결을 실선으로 해주다 보니 꺾임의 정도가 강하게 보여지고 있음
plt.show()
```

```python
# 3차 함수 f(x) = (x-2)x(x+2)

def f(x):
    return (x-2)*x*(x+2)

print(f(0))
print(f(-2))
print(f(2))
```

```python
# x값에 대해 ndarray 배열이면서 각각에 대응한 f를 한꺼번에 ndarray로 돌려줍니다.

print(f(np.array([1, 2, 3])))
print(type(f(np.array([1, 2, 3])))) #배열 타입으로 결과를 리턴해주고 있음
```

```python
# 그래프를 그리는 x의 범위를 -3 ~ 3까지로 하고, 간격 0.5 (간격이 좁을수록 부드러운 곡선의 형태로 그래프가 그려질 것임)

x = np.arange(-3, 3, 0.5)

plt.plot(x, f(x))
plt.show()
```

```python
# 그래프를 장식
def f2(x, w):
    return (x-w) * x * (x+2) # 함수 정의
```

```python
# x를 정의

x = np.linspace(-3, 3, 100) # x를 100분할하기
```

```python
# 차트 묘사
plt.plot(x, f2(x, 2), color='black', label='w=2')
plt.plot(x, f2(x, 1), color='blue', label='w=1')
plt.legend(loc="upper left") #범례위치를 위에 왼쪽으로 지정함
plt.ylim(-15, 15) # y축 범위
plt.title('(x-w)x(x+2)')
plt.xlabel('x') #x축에 출력해주고 싶은 값
plt.ylabel('f2(x, w)') #y축에 출력해주고 싶은 값
plt.grid(True) #눈금 표시
plt.show()
```

```python
# 그래프를 여러 개 보여주기

plt.figure(figsize=(10, 3)) # 전체 영역의 크기를 지정해주는 것
plt.subplots_adjust(wspace=0.5, hspace=0.5) # 그래프의 간격을 지정

for i in range(6): #6개의 이미지를 출력
    plt.subplot(2, 3, i+1) #2행 3열, 몇번째에 이미지를 넣어줄 것인지를 지정 / subplot의 카운팅은 1부터함
    plt.title(i+1) #인덱스에 해당하는 값으로 타이틀을 출력하도록 함
    plt.plot(x, f2(x, i))
    plt.ylim(-20, 20)
    plt.grid(True) #눈금도 표시
    
plt.show() 
#한 화면에서 여러개의 이미지를 출력하고 싶을 때 fiture로 영역을 지정해주고 subplots_adjust로 간격(인치의 단위)을 지정해주도록 함
```
