# XGBoost(eXtreme Grandient Boosting)
- 과적합 방지가 가능한 규제가 포함되어 있다.
- CART(Classification And Regression Tree)를 기반으로 하므로 분류와 회귀 둘 다 가능하다.
- 조기종료(early stopping)을 제공.
- 앙상블 부스팅의 특징인 가중치 부여를 경사 하강법(gradient descent)으로 함.
      
      
* 사이킷런에는 아직 이 알고리즘이 포함되어져 있지 않기 때문에 따로 설치를 해줘야 한다.

```python
# xgboost 설치
# 아나콘다 프롬프트 접속 -> conda activate tf_cpu -> conda install -c anaconda py-xgboost -> y
import xgboost

print(xgboost.__version__)
```

```python
# 위스콘신 유방암 예측
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import warnings 
warnings.filterwarnings('ignore') #주황색계통의 메시지들 출력을 하지 않도록 함.
```

```python
dataset = load_breast_cancer() #암환자에 대한 데이터를 읽어올 수 있음
X_features = dataset.data #훈련용 데이터
y_label = dataset.target  #정답 데이터

cancer_df = pd.DataFrame(data=X_features, columns=dataset.feature_names)

cancer_df['target'] = y_label # 1이면 암환자, 0이면 일반환자

cancer_df.head() #검진을 통해 주어진 값들을 출력해본 것임
```

```python
print(dataset.target_names)
print(cancer_df['target'].value_counts())
#malignant: 악성을 의미하는 암환자 / benign: 양성을 의미
```

```python
X_features = cancer_df.iloc[:, :-1] #-1 제일 끝 타겟항목 직전의 항목들을 선택 => 학습용 데이터를 꺼내옴
y_label = cancer_df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X_features, y_label, test_size=0.2, random_state=156) # 8:2 비율로 나누도록 함.

#전통적인 분리방법은 학습용 데이터에서 데이터를 학습과 검증으로 데이터를 나눠야 함
#위에서 만든 X_train, y_train을 다시 쪼개서 90%는 학습, 10%는 검증용 데이터로 분리.
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=156) #검증용데이터를 뽑아냄

print(X_train.shape, X_test.shape)
print(X_tr.shape, X_val.shape)
```

### 사이킷런 wrapper XGBoost의 적용

```python
# 사이킷런 wrapper XGBoost 클래스인 XGBClassifier 임포트
from xgboost import XGBClassifier 


xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3).fit(X_train, y_train) 
#경사하강법을 이해해야 learning_rate(학습주기로 생각)를 이해할 수 있음 -> 딥러닝과 관련된 회귀분석에 기본개념을 이해해야 함
```

```python
w_predicted = xgb_wrapper.predict(X_test)
w_pred_proba = xgb_wrapper.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score

def get_clf_eval(y_test, pred=None, pred_proba=None): #(정답, 모델이 예측한 값, 예측한 변수값의 확률)
    accuracy = accuracy_score(y_test, pred) 
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(
    accuracy, precision, recall, f1))
```

```python
get_clf_eval(y_test, w_predicted, w_pred_proba)
```

```python
# 조기 종료(early stopping)을 제공 
#=> 학습을 진행하면서 정확도가 떨어지는 시점이 오면 학습을 시키지 않고 최고점에서 학습 횟수가 남아있더라도 멈추도록 함

xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3)  
evals = [(X_test, y_test)]
xgb_wrapper.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='logloss', eval_set=evals, verbose=True) 
#학습을 시키도록 한다. / 조기종료는 early_stopping_rounds파라메타를 이용하는데 100을 넣어주도록 한다.
#                        함께 연동하여 사용되어지는 파라메타는 eval_metricf로 성능평가의 셋팅을 해주면 됨(logloss:손실 함수 등록) 
#                        +) eval_set파라메타도 추가적으로 등록 !!
#내부에서는 값을 체크하면서 정확도가 떨어져갈때 추가적으로 verbose의 설정값까지 체크하면서 
#조기종료시점을 찾아서 정확도 값이 떨어지면 더 이상 학습이 진행되지 않도록 한다.

ws100_preds = xgb_wrapper.predict(X_test) #최종적인 예측값을 반환
ws100_pred_proba = xgb_wrapper.predict_proba(X_test)[:, 1] #모든 행의 예측 정답 데이터를 가져와서 변수에 저장
#손실값을 메시지로 출력해주고 있음 => verbose를 True로 설정했기 때문임
#0의 근접한 값이 최적의 모델임! 진행될수록 값이 줄어들고 있지만 0.08--부터는 값의 변화가 거의 없음. -> 종료
```

```python
get_clf_eval(y_test, ws100_preds, ws100_pred_proba)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 12))
plot_importance(xgb_wrapper, ax=ax)
#30개의 핏쳐별 중요도에 대해 계산한 결과를 시각화하여 출력
```

```python
#[부스팅 앙상블: 약한 예측 모형(결과적으로 미흡한 데이터)들의 학습 에러에 가중치(데이터를 보강하는 것)를 두고, 
#                순차적으로 다음 학습 모델에 반영하여 강한 예측 모형을 만드는 것.]
# ***부스팅, 부스터: 하나의 모델에 학습을 시킴(부족한 데이터를 보강하여 다시 학습을 시키는 형식!!)
#    즉, 알고리즘은 하나인데 그 알고리즘을 데이터를 보완해가면서 학습을 시켜가는 것이다.(동일한 알고리즘에 여러 분류기를 만드는 과정으로 모델을 만듦)

# 배깅은 독립적으로 여러 개의 알고리즘을 생성하여 학습을 시키는 것

#부스팅은 순차적학습과 가중투표라는 개념이 반사적으로 떠올리도록 해야한다.
#순차적학습: 학습을 시키는데 학습 데이터가 분류적으로 결과가 미흡할 때는 그 미흡한 데이터를 가지고 다시 보강하는 방식!! 
#             그렇게 같은 모델을 가지고 순차적으로 학습을 시켜가는 방법을 말한다.

#현재 부스팅 중에 XG 부스트의 정확도가 상당히 높게 나와서 가장 많이 활용이 되어지고 있는 알고리즘 중에 하나 이다.

#가중 투표(테스트할 때 활용): 투표자의 능력치에 따라 한 표의 가치가 다른 투표를 의미한다.
# 회사에서 회식을 할 때 부장님이 말씀하신 장소로 결정됨 (같은 비중이 아닌 높은 능력치를 가진 자에 따라서 결정됨)

#순차적으로 데이터를 보강해가면서 동일한 알고리즘의 정확도를 개선시켰다고 한다면 
#최종적인 평가는 테스트 데이터를 집어넣었을 때 기존 정확도 값이 계산되어져서 나오게 될 것임
#이때 분류기를 통해서 값을 분류했다고 할 때 하드보팅은 (배깅같은 경우 다수결 원칙이였음) 
#     분류별로 동일건을 더해주고 높은 확률을 가진 값을 최종결론으로 판단한다.

# 소프트 보팅은 가중투표가 좀 더 명확해지는 최종판단의 분류를 하는 방법임
#  나오는 정확도(가중치)를 각 분류(확률)값을 곱하고 더해서(총합) 그 결과로 산출된 값에서 큰 값으로 분류하도록 한다.
```
