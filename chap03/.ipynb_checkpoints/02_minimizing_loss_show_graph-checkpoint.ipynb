```python
import numpy as np
import tensorflow as tf
```

```python
x_train = [1, 2, 3, 4]    
y_train = [0, -1, -2, -3]

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=1, units=1)) #세포 생성

sgd = tf.keras.optimizers.SGD(learning_rate=0.1) #모델에 사용할 설정정보
tf.model.compile(loss='mse', optimizer=sgd)
```

```python
import os 
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True' 
#디폴트는 False인데 True로 설정 => fit하여 학습하는 과정에서 계산되어진 훈련에 대한 loss값을 리턴해주게 된다.

loss = tf.model.fit(x_train, y_train, epochs=200)
```

```python
# loss 값의 시각화

import matplotlib.pyplot as plt

plt.plot(loss.history['loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left') #범례의 위치는 loc로 지정 (위 왼쪽에 배치)
plt.show()
#처음에는 랜덤한 값을 가지기 때문에 오차값이 크고, 5이하의 학습에서는 급격하게 손실이 줄어드는 것을 결과로 확인할 수 있다.
#시각화한 결과를 보면 200번이 아닌 20번만 해도 근사치값을 찾아갈 수 있다는 것을 알 수 있다.
```
