```python
#다중 선형 회귀 분석
#예시자료: 시험성적에서 기말고사 점수 예측
#기존 가설함수에서 입력데이터가 늘어나는 것임!! 
# => 펼쳐지는 수식은 매트릭스(행렬연산) 수식을 이용하면 모든 데이터를 한번에 계산해서 반환해줌
# ***각각의 기울기값들을 찾아내는 것이 최종 목표이다.

import tensorflow as tf
import numpy as np
```

```python
x_data = [[73, 80, 75],
          [93, 88, 93],
          [89, 91, 90],
          [96, 98, 100],
          [73, 66, 70]] #quiz1, quiz2, 중간

y_data = [[152], [185], [180], [196], [142]] #기말
```

```python
tf.model = tf.keras.Sequential() #신경망 객체 생성 

#다음으로는 세포를 추가(뉴런의 추가)
tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1))
tf.model.add(tf.keras.layers.Activation('linear')) #선형회귀분석을 활용하겠다는 명시적인 표현 / Activation default가 linear임.
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5)) #10의 5승분의 1
tf.model.summary()
```

```python
loss = tf.model.fit(x_data, y_data, epochs=200) #학습을 100번만 시켜보도록 함
```

```python
# loss 값의 시각화

import matplotlib.pyplot as plt

plt.plot(loss.history['loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left') #범례의 위치는 loc로 지정 (위 왼쪽에 배치)
plt.show()
```

```python
y_predict = tf.model.predict(np.array([[72, 93, 90]]))
y_predict
```
