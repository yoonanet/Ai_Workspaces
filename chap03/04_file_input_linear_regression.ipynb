```python
#주제선정 -> 데이터수집(파일의 형태로 담겨서 제공되어질 것임) 
# -> 파일에 있는 데이터를 읽어와서 처리할 때 어떻게 처리하면 되는지에 대한 실습
import tensorflow as tf
import numpy as np
```

```python
data = np.loadtxt('../data/data-01-test-score.csv', delimiter=',', dtype=np.float32) 
#테이블 형태로 데이터가 저장만 되어져 있다면 다 읽어올 수 있음 / delimiter:구분자의 의미를 가진 파라메타

x_data = data[:, 0:-1]
y_data = data[:, [-1]] #마지막 항목만 지정하려면 리스트로 -1을 지정하도록 한다.
#선형회귀는 대표적인 지도학습임. 따라서 정답으로 사용할 데이터를 분리시키도록 한다.
```

```python
print(x_data, '\nx_data shape:', x_data.shape)
print(y_data, '\ny_data shape:', y_data.shape)
#리스트형태로 데이터를 반환해주고 있음.
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1, activation='linear')) #가장 선호하는 코드(가독성도 높이고 간단하게 작업할 수 있는 코드) 
#input_dim:입력데이터 핏쳐 갯수, units:출력데이터 핏쳐 갯수 
#tf.model.add(tf.keras.layers.Activation('linear')) 가독성을 위해 붙인 것임
```

```python
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5)) #learning_rate는 디폴트가 0.1부터임
```

```python
tf.model.fit(x_data, y_data, epochs=2000) 
#learning_rate의 값을 작게 넣으면 최저점까지 도달할 수 있도록 학습값을 높게 잡아줘야 함
```

```python
#세 개의 w값 출력하기
tf.model.layers[0].get_weights()
#현 데어터의 오차가 최소화되어지는 w와 b의 찾은 값을 출력
```

```python
tf.model.predict([[60., 70., 80.], [73., 80., 75.]])
#단순회귀분석: 최소의 오차값으로 w의 값을 찾는 것
#예측 데이터로 이미 학습을 시킨 데이터를 집어넣었을 때 정확도가 100%로 나올 것이라는 착각을 할 수 있음 => (정답 152.으로 결과가 나올까?)
# -> 예측 모델은 데이터들 중에서 오차가 최소가 되는 직선 하나를 찾은 것이기 때문에 실제 데이터와의 연관성으로 결과가 나오는 것이 아님.
```
