```python
#데이터에서 각 핏쳐들간에 단위가 다를 때(학습이 안됨) 표준화를 해줬었음. 이에 대한 실습 진행
import tensorflow as tf
import numpy as np
```

```python
# open(시작가), high(최고가), volume(거래량), low(하한가), close(종가)
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
def min_max_scaler(data): #정규화 기능 구현
    numerator = data - np.min(data, axis=0) #담겨진 데이터에서 행 기준의 최소값을 찾아서 빼줌
    # 최소값인 데이터는 0이 될 것이다. 자기자신을 빼기 때문임!! 
    denominator = np.max(data, axis=0) - np.min(data, axis=0) #각각의 핏쳐에서 행의 최대값과 최소값의 차를 저장하게 됨.
    return numerator / denominator #계산된 값을 나눠줌 -> 0과 1사이의 값으로 범위를 변환한 것임.
#내 값이 가장 크면 1로 만들겠다는 것임. 내 값에서 가장 작으면 0으로 만들겠다는 것임.
#즉, 정규화라는 것은 가장 큰 값을 1로, 가장 작은 값을 0으로 나머지 값들은 0과 1 사이에 값으로 매칭시키도록 하겠다는 것이다.
#단위가 다르다고 하더라도 정규화를 하면 0에서 1사이 값으로 공통된 값의 형식으로 표현이 되는 것이다.
#- 표준화는 데이터 범위의 값을 평균을 0으로 표준편차를 1로 바꿔주는 것


#주성분분석에서는 표준화를 했을 때 우리가 원하는 데이터가 나온다는 것이고, 
#일반데이터에서는 정규화를 시켜주면 단위에 일치성을 가져가는 효과를 가져감
```

```python
data_scaler = min_max_scaler(xy)
data_scaler
```

```python
x_data = data_scaler[:, :-1]
y_data = data_scaler[:, [-1]]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1],
                                   units=y_data.shape[1],
                                   activation='linear'))
```

```python
tf.model.compile(loss='mse',
                 optimizer=tf.keras.optimizers.SGD(learning_rate=1e-1))
```

```python
loss = tf.model.fit(x_data, y_data, epochs=2000) #발산을 안하는 것은 좋았지만 값의 변화가 거의 없음.
```

```python
print(loss.history['loss'])
```

```python
predictions = tf.model.predict(x_data)
```

```python
score = tf.model.evaluate(x_data, y_data)
```

```python
print('Prediction: \n', predictions) #정규화된 값으로 결과를 출력되고 있음
print('Cost : ', score) 
```

```python
#역 정규화의 과정으로 값을 원래의 값으로 변환 -> 값이 100% 똑같게 변환될 수 없다는 것을 알고 있어야 한다.
```
