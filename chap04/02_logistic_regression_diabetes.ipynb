```python
import tensorflow as tf
import numpy as np
```

```python
data = np.loadtxt('../data/data-03-diabetes.csv', delimiter=',', dtype=np.float32) #환자에 대한 검사결과가 담긴 데이터 파일임

x_data = data[:,:-1]
y_data = data[:,[-1]] #마지막 컬럼항목만 가지고 오도록 함.
```

```python
print(x_data.shape, y_data.shape)
```

```python
tf.model = tf.keras.Sequential() #신경망 생성 (머리통만듦)

tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1], units=y_data.shape[1], activation='sigmoid')) #머리통 안에 세포들 구성
#여러가지 요소의 세포들을 layers안에 Dense의 클래스 이름으로 제공을 해주고 있음
#8개의 데이터로 정확도가 높게 나오지 않을 때는 유사한 성격을 띄는 핏쳐들을 찾으면서 차원축소등을 분석의 전처리를 수행하게 됨. 
#-> 따라서 input_dim을 8개로 지정하게 되면 바뀔 수 있음. 코드가 자동으로 변화된 것으로 수정하고 싶다면 
#                                                        x_data.shape의 열의 갯수 데이터를 가지고 오도록 한다.

tf.model.compile(loss='binary_crossentropy', #내부적으로 교차적으로 검증을 진행하고 있는 것임.
                 optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
                 metrics=['accuracy'])

tf.model.summary()
```

```python
loss = tf.model.fit(x_data, y_data, epochs=500)
#loss는 잔차 -> 이상적인 잔차값은 0
#결과적으로 보면 오차값은 최소화되면서 결과값은 높아지는 것을 확인할 수 있다.
```

```python
loss.history['accuracy'][-1]
```

```python
#지금 현재 신경망 알고리즘에서 정확도가 높게 나오지 않았다면 상관분석 알고리즘을 통해서 유사성이 높은 필드들을 추출해내도록 한다.
tf.model.predict([[-0.882353,-0.0653266,0.147541,-0.373737,0,-0.0938897,-0.797609,-0.933333]]) #0.5보다 크기 때문에 true로 판단한 것임
```

```python
# 검증과 관련된 함수 ↓ (전체데이터에 대해서 검증을 하도록 한다.)
evaluate = tf.model.evaluate(x_data, y_data)

print('loss:{0}, accuracy:{1}'.format(evaluate[0], evaluate[1]))
#검증에 의해서 도출되어지는 최종값이 위에랑 일치하게 나오고 있음.
```
