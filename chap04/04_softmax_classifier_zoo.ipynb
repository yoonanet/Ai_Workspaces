```python
import tensorflow as tf
import numpy as np

data = np.loadtxt('../data/data-04-zoo.csv', delimiter=',', dtype=np.float32)

x_data = data[:,:-1]
y_data = data[:,[-1]]

print(x_data.shape, y_data.shape)
```

```python
nb_classes = 7 #분류의 갯수 = 7의 sigmoid => 여러개의 분류군을 가진 것을 softmax
#실제 데이터의 결과값은 범주군이 숫자로 매핑되어져 있음 이때, 원합인코딩방식으로 데이터를 펼쳐줘야한다.

#유용한 기능들을 제공해주는 utils안에 원합인코딩이 가능한 클래스가 있음
y_one_hot = tf.keras.utils.to_categorical(y_data, nb_classes) #정답데이터와 몇개의 분류로 지정이 되어져 있는지를 넣어주면 됨
y_one_hot #결과적으로 분류에서 숫자로 매핑된 값에 1로 표시해주면서 데이터를 바꿔서 리턴된다는 것을 확인할 수 있다.
          #즉, 각각의 정답데이터를 따로 바꿀 필요가 없는 것이다.
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1], units=nb_classes, activation='softmax'))
```

```python
tf.model.compile(loss='categorical_crossentropy',
                 optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
                 metrics=['accuracy'])
```

```python
tf.model.fit(x_data, y_one_hot, epochs=1000)
```

```python
test_data = np.array([[1,0,1,0,1,0,0,0,0,1,0,0,6,0,0,0]]) # 5

pred = tf.model.predict(test_data)
pred
```
