```python
#신경망 알고리즘 => 하드웨어 성능이 떨어져서 1차 암흑기
#태생에 기원이 되는 이론이 퍼셉트론으로부터 파생되어진 이론이다.
#퍼셉트론은 프랑크 로젠블라트 학자가 고안하였고, 딥러닝(신경망)의 기원이 되는 알고리즘이다.
#           인공뉴런(더 많이 불려짐.)이나 단순퍼셉트론이라고 불러주고 있음
#           두 개의 신호(핏쳐)가 있을 때 신경에 선형회귀는 가설함수를 H(x) = Wx + b -> 입력신호가 하나 일 때임. 
#                                                                   / H(x1, x2) = w1x1 + w2x2 + b -> 입력신호가 두 개일 때 

#임계값은 전압이 오는데 5볼트의 전압에서 2볼트 이하가 전압이 오면 신호를 내보내지 않겠다는 것임. 2볼트 이상일때 신호를 보내겠다는 것
#그때 기준으로 잡은 2볼트가 theta(임계값)라는 것이다. => 뉴런의 활성화를 시켜주는 기준값이 된다.

#cpu의 내부를 살펴보면 논리연산자(and, or, not)로 설계가 되어져 있음.
#논리연산자는 값을 true(신호를 보냄)와 false(신호를 보내지 않음)로 표현하고 있었음.
#             and게이트라면 신호가 둘 다 들어왔을 때 신호를 출력. or게이트는 하나의 신호라도 들어오면 출력을 내보냄.

# 신경망 알고리즘이 두 개의 암흑기를 거치면서 신경망 자체는 더이상 발전을 할 수 없다는 
# 인식이 잡히다 보니 이슈가 되어지던 자식 개념의 용어인 딥러닝이라는 이름으로 불려지고 있는 것임 
#즉, 딥러닝 = 신경망으로 같은 의미로 생각해도 됨.
#딥하게 학습을 시키겠다는 것으로 딥의 의미가 레이어를 여러개 만들겠다는 개념으로 딥러닝의 이름이 지어짐.

#XOR게이트: 실질적으로 제일 많이 탑제되어짐 / 서로 부호가 다를 경우에만 1을 내보내고, 같을 경우에는 0을 내보내는 소자
#           이는 데이터를 나눌 수 없음 따라서 ai의 2차 암흑기임
# 곡선으로 그리면 나눌 수 있음(특정영역에서 일정 값을 통과 시키면 직선이 곡선의 효과를 가질 수 있음) 
# -> 이는 sigmoid함수를 통해서 통과시키도록 하면 됨. 또한 단독으로 게이트를 구성하는 것이 아닌 게이트를 조합해서 표현하도록 함. => 실제로 계산하면 xor을 구성할 수 있음

#다층 퍼셉트론 -> 여러 레이어로 구성하는 것 -> 신경망 -> 딥러닝(신경망에 일부 개념)
```

```python
import tensorflow as tf
import numpy as np
```

```python
# Gate

x_data = np.array([[0,0], [0,1], [1,0], [1,1]]) #신호에 대한 경우의 수를 전부 x_data에 담아주도록 한다.
# y_data = np.array([[0], [1], [1], [1]]) # OR Gate / 정답데이터
#y_data = np.array([[0], [0], [0], [1]]) # AND Gate

y_data = np.array([[0], [1], [1], [0]]) # X-OR Gate / 다층 퍼셉트론으로 구성
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=1, input_dim=2, activation='sigmoid'))
#Dense의 파라메타 순서는 units가 먼저임 / 지금의 예측값은 바이너리의 값임. 따라서 sigmoid를 적어주도록 한다, 
#                                         3개 이상의 분류는 softmax, 연속적인 값을 예측할 때는 linear

tf.model.compile(loss='binary_crossentropy', 
                 optimizer=tf.optimizers.SGD(learning_rate=0.01),
                 metrics=['accuracy']) #분류분석은 성능평가의 지표까지도 보여주게끔 요청할 수 있음

tf.model.summary()
#여기서 파라메타의 값에 대한 정보가 3개로 보여주는데 2개는 입력값, 1개는 바이어스값(b)을 말하는 것이다.
```

```python
loss = tf.model.fit(x_data, y_data, epochs=4000)
```
