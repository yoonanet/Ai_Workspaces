```python
import tensorflow as tf
import numpy as np

# XOR Gate
x_data = np.array([[0,0], [0,1], [1,0], [1,1]]) 
y_data = np.array([[0], [1], [1], [0]]) # X-OR Gate 
```

```python
#딥러닝은 어떻게 확장되느냐
#dense를 여러 개 구성할 수 있는데 입력은 핏쳐로 고정되어져 있음 
#출력은 갯수를 여러 개 할 수 있음. dense(은닉층)은 우리가 추가를 해가는 것이기 때문에 갯수는 제한이 없다. (+)출력의 갯수를 우리가 임의로 정의할 수 있음) 
#딥러닝에서는 1024나 512로 출력의 갯수를 지정해놓고 있음
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=10, input_dim=x_data.shape[1], activation='sigmoid')) #첫번째 dense에는 꼭 입력 갯수 넣어줘야 함
#은닉층 4개 추가
tf.model.add(tf.keras.layers.Dense(units=10, activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
tf.model.compile(loss='binary_crossentropy',
                 optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
                 metrics=['accuracy']) #베니싱의 문제를 해결할 수 있는 알고리즘이 Adam임
```

```python
tf.model.summary() # 2 * 512 + 바이어스 -> 첫번째 param
```

```python
loss = tf.model.fit(x_data, y_data, epochs=20000)
#딥러닝을 했는데 예측률이 반밖에 안됨. 왜이럴까 sigmoid를 적용하면서 배니싱에 빠져서 그러는 것임.
#정확도가 떨어진다면 추가적으로 체크할 부분이 있음. 
#선형회귀는 2차함수로 결과값은 한 개임. 근데 sigmoid를 적용하게 되면 다양한 굴곡이 생길 수 있음 이때 정확도 부분에서 배니싱에 빠짐

#Adam을 적용해도 학습의 양이 부족해서 개선이 되지 않고 있음
#지금 현재 과대적합에 빠져서 더이상 개선이 안되고 있는 경우임.
```
