```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data() #차원을 읽어오면 2차원, 데이터는 6만개

#학습을 시키기 전에 정규화
# reshape에서 -1을 넣으면 가변을 의미 -> train_input의 데이터 갯수만큼 자동으로 잡아주게 됨
train_scaled = train_input.reshape(-1, 28, 28, 1) / 255.0 
#정규화는 나의 최대값 빼기 최소값을 한 값으로 나눠주도록 함 / 최소값이 0이기 때문에 뺄셈을 해줄 필요가 없음
#이미지 데이터는 기본적으로 3차원임 /  6만개의 데이터를 고려하게 된다면 4차원까지 고려해야함 
#                                     (이미지 본연의 행렬, 채널, 레코드데이터까지 같이 고려해야하기 때문)
# 텐서플로우나 케라스, 파이토치의 라이브러리 패키지들에서는 채널의 표현방식을 last채널을 사용함. -> (60,000, 28, 28, 1)
# 입력데이터를 4차원으로 변환하면서 정규화하고 리턴해주게끔 변수를 지정 (4차원의 6만개 데이터가 저장됨)

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=1234) 
#학습데이터를 80%는 훈련데이터, 20%는 검증데이터로 학습 전에 나눠주도록 한다.
```

```python
# CNN(합성곱) 신경망 만들기

model = keras.Sequential() 

# 합성곱 연산
model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), padding='same', input_shape=(28, 28, 1), activation='relu')) 
#합성곱은 Conv2D클래스로 제공해주고 있음. -> 이클래스가 합성곱의 연산을 처리하도록 정의되어져 있음
#임의로 filter의 갯수를 지정하면 filter마다 고유 특징들을 가지게 될 것임 (filter가 많을 수록 정밀하게 특징을 가져가게 됨)
#filter를 셋팅할 때  kernel_size파라메타를 활용 따라서 CNN에서 둘 다 같은 용어로 사용함
#padding(더미셀)에 지정할 수 있는 값은 valid벨리드랑 same(입력값과 동일하게 출력하겠다는 의미)로 지정이 가능함
#출력의 활성화함수를 relu로 지정해주는 것이 일반적임.

#합성곱과 활성화함수까지 거쳐서 나온 결과 데이터를 핏쳐맵이라고 불러주고 있음
#이렇게 나온 핏쳐맵을 풀링에 통과시키면서 데이터의 특징을 부각시키도록 한다. (max풀링을 적용하면 2바이2에서 최대값을 뽑아냄)
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
#pool_size는 일반적으로 이 값은 보여주기 위한 셋팅이고, 디폴트값을 사용하는 것이 일반적임 (2바이2) 
# => 4가지의 값에서 하나의 값을 꺼내오는 방식

#일반적으로 딥러닝에서 보면 합성곱/ 풀링계층망을 많이 가져갈 수록 정확도는 높아짐
```

```python
#합성곱 추가
model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu')) 
#두번째부터는 input_shape은 지정해주지 않아도 됨. 
#첫번째 합성곱에서 이미 shape이 지정이 되어 구성이 되기 때문에 다시 적용하게 되면 에러날 수 있음 
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

#이전까지는 데이터가 행렬로 구성을 하고 있을 것이고 이를 잘 담아서 고유의 값이 필터마다 학습할 때마다 최적의 값으로 셋팅될 것임
#마지막은 일반 신경망으로 연결해주면 됨 -> 어파인계층(선형대수)
```

```python
# 완전 연결형(Affine layers) - 신경망 연결
#지금 현재 shape은 4차원을 띄고 있을 것이고, 이때 신경망에서는 1차원으로 쭉 펼쳐서 입력으로 넣어줘야 할 것임

model.add(keras.layers.Flatten()) #Flatten: 1차원으로 쭉 펼쳐주게끔 제공되어지는 클래스 / 연결형.

model.add(keras.layers.Dense(units=100, activation='relu')) #은닉층으로 100개정도 출력이 되도록 지정. 은닉층같은 경우 활성화함수를 relu로 지정
# 입력갯수(input_dim)는 신경안써도됨. Flatten으로 입력값이 펼쳐지면 자동으로 넣어짐(28 * 28 * 64)
# 여기서도 여러 개의 은닉층을 만들 수 있지만 한개만 만들어줌

model.add(keras.layers.Dropout(0.4)) #과적합에 빠지지 않도록 적용하는 기술을 활용 => Dropout 
#-> 학습과정에서 100개를 다 사용하는 것이 아닌 랜덤하게 활성화 및 비활성화하여 학습을 시키도록 함
#    40%를 드롭시키고 60%로 학습을 시키도록 지정을 해주도록 한다.

#마지막으로 소프트맥스를 거쳐서 최종결과를 도출하도록 한다. (확률적인 값을 추출하기 위함)
model.add(keras.layers.Dense(units=10, activation='softmax')) #분류군이 10개이기 때문에 고정된 값을 지정하도록 한다.
```

```python
model.summary()

#CNN에서는 W가 필터임. x가 입력 데이터임 => X * W + b (신경망에서는 784+10+10으로 7850이 나올 것임)

#첫번째 320의 파라메타 값이 나오는 이유: 3바이3의 필터로 9개가 만들어질 것임.
# 필터 갯수는 32일 것이고 32 * 9 * 1(채널) => 288 
# 바이어스는 출력의 갯수 필터로 32개 잡았음 
# 여기서 288 + 32 = 320으로 결과를 보여준 것임 

#두번째 합성곱 계산 ↓
#3바이3으로 하나는 9개일 것임. 필터 64개임
#두번째의 채널 의미는 출력의 갯수로 32만큼 나올 것임
#그렇게 되면 64 * 9 * 32 + 64로 계산하면 되는 것임
```

```python
#Affine(선형대수) 연결
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
history = model.fit(train_scaled, train_target, epochs=50, batch_size=100, validation_data=(val_scaled, val_target)) 
#validation_data 파라메타를 통해서 학습시킬때 검증용 데이터를 넣어주면서 정확도도 같이 확인 가능
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss']) #손실의 값을 시각화
plt.plot(history.history['val_loss']) #검증의 손실값

plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val']) #범례 타이틀 지정
plt.show()
#손실율이 학습을 할 수록 더 커지는 추세로써 과대적합이 일어난 것
```

```python
#딥러닝: 기존 신경망에 은닉층(레이어)을 여러 개 더 깊게 구성해서 정확도 높은 모델들을 만듦 (심층 신경망)

#딥하게 레이어층을 깊게 가져가면 더 높은 정확도를 가져갈까 
# 은닉층의 갯수가 증가되면 그만큼 w의 갯수가 증가되는 것이고 w마다 입력데이터에 대한 출력의 정답이 있을 때 
# w가 디테일한 정보들을 고유하게 가져가고 있으므로 정확도가 높아지는 것으로 예측

#대표적 딥러닝 알고리즘의 이미지 처리에 최적화되어진 합성곱 신경망(CNN)
# 90년대와 2000년때까지는 컴퓨터비전이라는 학문의 영역으로 연구가 진행되어졌었던 영역이였음
# 이 영역이 사람이 연구해서 이미지 데이터를 처리하는 기술이 ai가 높은 정확도에서 앞서가고 있음

#신경망 내에서도 이미지 처리를 위한 신경망을 CNN이라고 불러주고 있는 것임.
#어파인 계층에서 선형대수를 계산해주는 행렬의 내적 연산을 해주고 은닉층일 때 활성화함수를 relu로 연결해서 딥하게 연결해가는 것이 딥러닝이였음

#CNN(Convolutional Neural Network)에서는 Conv, relu, pooling들의 계층을 통해서 이미지를 처리하도록 하는 것이다.
#cunv는 이미지라는 데이터에 최적화되어진 개념, 이미지 데이터는 기존에 입력데이터와 어떤 특징이 있길래 cunv개념을 별도로 정의하고 있는지 파악해야함
# 기존에는 1차원배열로 펼쳐서 픽셀단위의 이미지를 처리했었음. -> 입력의 뉴런으로 구성해서 처리한 것임

#이미지의 특징은 배경과 특정된 영역이 공통된 색상을 가지고 있음. 
#기존에 행렬의 특징을 가진 데이터에서 1차원배열로 펼쳤을 때는 그 특징을 그대로 포함하고 있지 않다는 것
#무조건 펼치는 것이 아닌 이미지가 가진 특징 자체를 입력으로 집어넣어서 특징을 학습시켜보자는 특징의 이미지 알고리즘이 CNN이다.

#MNIST 데이터셋의 주 특징은 흑백이미지이고, 28 * 28의 크기를 갖는 데이터셋
#=> 백터화 시켜서 학습, 이미지 고유특징인 형상을 무시하고 모든 입력데이터를 동등한 뉴런으로 취급하여 
#   shape이 가지는 의미 정보를 살릴 수가 없는 학습법이였음

# 합성곱 계층: 3차원 데이터로 입력을 받으면 다음계층에서도 3차원데이터로 전달받으면 
#              이미지의 고유 특징을 그대로 살리면서 학습을 가질 가능성이 커짐

# 입출력계층을 핏쳐맵이라고 함
# Convolutional(합성곱): 이름에 곱셈 개념의 의미가 담김_ 연산이라는 것을 매칭하면 됨 
#  -> 이미지 데이터의 주 특징은 그 shape을 그대로 유지하면서 학습하고자 하는데 별도의 연산과 관련된 알고리즘을 제안하는 개념임
#  합성곱 연산: 입력데이터가 들어올 때 엣지를 통해서 w를 정의했듯이 입력데이터가 2차원의 형태로 전달이 되어진다면
# 연산할 수 있는 필터(오차를 최소화하자는 개념이 그대로 적용, 데이터에 따른 차원 정의)라는 것을 정의해서 데이터가 가지는 입력의 고유특징을 살리도록 함.

# => 처음에는 필터에 랜덤한 값이 담기게 됨. -> 그 필터 행렬에 맞춰서 기존 핏쳐 데이터에 곱셈 후에 덧셈을 진행 
#    -> 필터를 한칸 이동시킴 -> 그 다음 열로 이동됨 -> 계산 후 한 칸 아래로 이동 -> 각각 곱한 후 더하는 작업 진행
#      !!계산을 하고나면 데이터가 축소되어짐!!
# 이미지 처리에서의 필터 크기는 우리가 정해주면 됨. shape => 기본적으로 홀수 단위로 입력값보다 작은 값으로 정사각형의 형태로 필터가 정의됨
# 기존 신경망에서는 입력과 출력값에 따라서 고정값이 생기게 됨 


# 입력 데이터에 따라서 기존 신경망은 입력데이터 특징에 따라서 출력을 어떻게 내보내는지가 결정되면 자연스럽게 w갯수가 생성됐는데
# 그 개념을 필터로 그대로 가져오게 된 것임. 필터에 각 데이터들을 w로 생각하면 됨. (입력과 필터가 매트릭스)
# 단지, 입력데이터에 대해서 선형대수인 내적의 계산을 했었음. 이미지는 합성곱 연산법으로 계산을 함.

# 입력의 뉴런을 구성하고 핏쳐(입력)의 갯수만큼 Dense를 구성해서 Dense의 input_dim을 통해서 입력 전달 뉴런 생성, 
# units를 통해서 다음 은닉층에 몇개의 출력으로 내보낼지를 지정 (입력과 출력에서 연결되어지는 선을 엣지(w값)라고 표현)
# 신경망의 특징은 핏쳐와 출력갯수를 곱하면 w의 값을 추출할 수 있음.

#원래 합성곱의 연산에서는 바이어스값을 추가해서 최종값을 출력하는데 계산상에 편의성이나 알고리즘의 이해를 위해서 잠깐 보류

# 필터의 크기에 따라서 원데이터보다 축소가 되어짐. 원이미지 데이터의 크기를 그대로 유지하기 위해서 
# 원래 데이터(4바이4)의 바깥영역을 확장(양쪽끝에 공픽셀을 추가)해서 연산을 진행하도록 한다.
# -> 핏셀 하나씩을 더미(특정값으로 채움_ex.0)로 추가하여 연산을 진행한다면 원래의 4바이4크기로 출력이 되어질 것임
#   이때 더미로 추가해주는 픽셀을 패딩이라고 불러주고 있음

#필터 = 커널, 패딩_자동으로 데이터의 shape의 크기가 줄어들지 않고 유지할 목적으로 추가하는 셀(입력과 출력이 동일한 크기임.) -> 실무에서도 일반적으로 활용하는 방법
#스트라이드: 이동의 크기를 말함, 필터계산에서 사용되는 단위 / 실습에서는 한 칸씩 이동했었음 (default가 1임)
#!!일반적으로 3바이3, 5바이5로 필터를 정의함!!

# 컬러이미지는 RGB의 색상 표현단위로 컬러에 대한 처리를 데이터로 담아서 표현을 해주게 되어져 있음
# -> RGB의 색상을 채널이라고 불러주고 있음. (3차원으로 처리됨)
# 흑백이미지는 1채널을 가지고, 컬러이미지는 3채널을 가짐
# 3차원은 컬러데이터를 cnn에서 어떻게 처리하는지에 대한 개념: 
# ***채널이 3개면 합성곱을 하기 위해서 각 채널별 필터도 채널과 동일한 갯수를 반드시 가지고 있어야 함

#RGB의 순서로 담긴 데이터에 필터를 매핑해서 합성곱의 계산을 진행 -> 계산하면 하나의 값만이 나오게 됨 (염두할 두번째 사항)
#   3차원의 합성곱을 연산하려면 채널수에 동일한 갯수로 필터가 정의되어야 하고,  (염두할 사항1)
#   그렇게 해서 3차원의 합성곱을 계산하고 나면 하나의 값으로 결과가 도출된다.   (염두할 사항2)
#   _채널별 계산되어진 하나의 이미지 데이터가 하나의 값으로 도출되는 것 (매트릭스로 출력)

#필터 혹은 커널이라는 것을 정의해서 합성곱 연산을 해서 해당 입력데이터에 대한 학습을 진행하면서 
#이미지가 가지는 특징을 필터로 최적화해서 데이터가 가지는 고유 특징을 최종 업데이트 => cnn

#합성곱의 연산을 할 때는 누적의 개념으로 계산을 함. 이때 셀별로 계산한다면 1행1열은 한번씩으로 밖에 계산이 안됨
#데이터에 대한 소실, 중요정보가 많이 사라지게 될 수 있는 문제점이 발생하는 것임.
# 이를 방지하기 위해서 입력의 shape과 동일하게 출력하기 위해서 패딩의 방법을 활용

#행과 열의 값을 가지면서 RGB 댑스 개념을 가져가야해서 이미지는 3차원으로 처리되는 것임.

#입력 데이터(C, H, W) * 필터(C, FH, FW) = 1, OH, OW - 3차원
# first방식: 채널을 제일먼저 표현하는 방법, last: 채널을 가장 끝에 표현하는 방법 
# 채널을 기준으로 표현하는 방식을 의미, 텐서플로우에서는 디폴트로 last방식을 사용함

#데이터분석에서 입력으로 데이터 신호가 들어오고 1차적으로 이미지에 어떤 특징을 가지고 있는지를 필터를 통해서 찾아내도록 하는 것임.
# -> 하나만 있으면 눈에 있는 특징만 추출하게 되면 모두를 필터링을 하다보니 일부는 필터링을 못하거나 금방 오염돼서 갈아줘야하는 문제 발생.
# 따라서 여러 필터를 구성해서 큰 특징을 일차적으로 뽑아내고 그 다음 특징을 뽑아내는 개념으로 동작시키도록 해야한다.
# ***필터의 갯수가 많아질수록 입력에 데이터를 좀 더 정밀하게 특징들을 가져갈 수 있는 것이다. 

#하나의 이미지에 하나의 필터를 가져가는 경우는 거의 없음. 보통은 필터를 n개 구성(n개의 필터는 임의로 지정해줄 수 있음.)
#필터를 n개로 구성하게 되면 입력데이터에 대한 다른 특징들을 정밀하게 보관하게 될 것임.
# 필터가 2개였을 때는 28바이28로 그계산된 필터의 최종 결과가 하나 더 추가됨_ (2, 28, 28)로 결과가 나오게 됨
# 즉, 필터의 계산을 n개로 한다면 계산해서 도출되는 값들이 n개가 됨.          (n, 28, 28) => 필터의 갯수만큼 결과값이 출력으로 나옴

#원데이터에 필터가 늘어나게 되면 데이터는 누적되면서 데이터가 뚱뚱해짐 
# => 하나의 이미지인데 필터가 많아지면 하나의 이미지가 여러개 n개로 구성되면서 이미지의 특징이 부각되어짐!!

#이미지의 입력데이터는 3차원이고, 그렇게 걸러내는 필터도 3차원이였음
#***하지만 필터를 n개로 구성하게 되면 4차원임 / ***최종적으로 바이어스를 추가적으로 더해서 계산해주면 됨.

#기존에는 한개의 입력데이터로 예시를 봤었음. 
#MNIST의 훈련데이터로 6만개를 제공해주고 있음(신경망에서는 모두 학습을 시키는 것이 아닌 미니배치로 랜덤하게 일부만 가지고 와서 학습을 시킴)
# 훈련 48,000개, 검증 12,000개로 나눠서 학습을 시키도록 함
# 기존 한개의 입력이었을 때는 3차원이였지만 n개의 배치 학습을 시키려면 4차원으로 확대가 되어진다.

#4차원으로 표현할 방법이 없기 때문에 3차원에 n개의 데이터를 추가적으로 표시를 해주고 있는 것임.
# n개의 이미지를 입력받게 되면 4차원의 입력데이터를 받아서 4차원의 필터를 전달받아서 4차원의 결과를 얻어낸다.

#(데이터갯수, 채널수, 넓이, 높이)의 4차원으로 데이터를 넣도록 함.
#ex) 입력데이터배열표현(100, 1, 28, 28) / 필터배열표현(20, 1, 3, 3) / 바이어스배열표현(20, 1, 1) / 결과배열표현(100, 20, 28, 28) - 패딩을 통해 입력값과 동일하게 했을 때,
# **b자체는 2차원이고, 필터갯수만큼 바이어스가 생기고, b는 상수라서 1로 지정

#***CNN에서는 각 필터별 값에 특징을 규정짓는 w의 값을 찾아내는 것이 핵심이다.

#CNN알고리즘에서 풀링계층 용어를 이해해야함
# 풀링: 데이터 축소라고 보면 됨.
# 앞에서 패딩을 적용하지 않으면 자동으로 출력이 데이터 축소가 일어나고 있었음.
# 일반적으로 합성곱을 계산할 때는 패딩을 적용을 해서 데이터의 축소를 만들지 않음.

# ***그렇다면 풀링이 왜 필요한 것일까? (기본적으로 합성곱의 결과로는 데이터가 축소되는데 풀링에서 차원을 왜 축소하는가)
# 합성곱에 의해서 데이터가 축소되는 것보다 합성곱을 수행한 다음에 통과되어진 데이터를 활성화함수를 거쳐서 
# 최종출력이 나오면 그 뒤에 데이터를 축소하는 것이 훨씬 더 좋은 결과가 도출된다는 것을 발견함
# 알고리즘을 수행 후 풀링이라는 개념을 별도로 추가해서 합성곱의 신경망에 따로 적용하는 것임

# 패딩을 하는 이유
# 1. 가장자리에 있는 데이터가 한번 밖에 사용이 안되니까 모서리 부분에 중요정보가 있다면 손실됨 -> 정보를 그대로 살리는 것이 목적
# 2. 패딩을 하지 않고 데이터를 축소시키면 최종 출력에 좋은 결과로 도출이 되어지지 않음

#풀링은 어떻게 데이터를 축소하는가?
#합성곱에 의해서 거쳐서 나온 데이터 결과를 내보낼 때 활성화함수를 통해서 내보냄. 이때 사용하는 활성화함수가 렐로함수.(성능좋은 모델을 만듦)
#바로 은닉층에 내보내는 것이 아니라 풀링이라는 계층(데이터 축소)을 추가해서 내보냄. 

#데이터 특징을 부각시키려는 것 / 상대적으로 가장 높은 특징을 더 부각시킴

# 풀링은 두가지의 종류가 있음 - max풀링, average풀링
# 최대값풀링: 4개에서 최대값 하나 뽑아냄, 평균풀링: 4가지의 값을 평균내서 뽑아냄
#일반적으로 풀링은 2바2로 잡음 /  max풀링 적용 -> 그때 잡은 풀링에서 2번 스트라이드하겠다는 것 -> 풀링에서 가장 큰 값을 뽑아내어 값을 넣음 
# => max풀링을 일반적으로 사용함 (특징에 대한 부각이 더 잘됨. 평균풀링을 사용하면 고유 특징이 상쇄될 수 있음)

# CNN목적: 합성곱 -> 렐루 -> 풀링은 여러번 적용 가능, 몇 번의 적용을 통해 최적을 찾아낼 수 있는지를 추적해내야 함. 
# 최종적으로는 분류를 하고 싶은 것 ->  최종의 분류는 마지막 단계에서 어파인의 신경망을 그대로 붙여주면 됨
# 현재 신경망안에 합성곱만 들어가있으면 CNN이라고 불러주는 것임.

#신경망 구조
#데이터를 일반신경망에 적용하기 위해 1차원배열로 펼침. -> Affine(선형대수연산 = 행렬의 내적 H(x) = wx + b) - 뉴런 구성 10개, 748바이10 W가 구성됨 
#CNN구조 합성곱연산
```
